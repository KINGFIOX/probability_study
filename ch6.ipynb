```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 生成自由度为10的卡方分布随机变量
x = np.linspace(0, 40, 100)
y = stats.chi2.pdf(x, df=10)

# 设置图像标题和坐标轴标签
plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', lw=2)
plt.title('Chi-Square Distribution with df=10')
plt.xlabel('x')
plt.ylabel('CDF')

# 显示图像
plt.show()
```
