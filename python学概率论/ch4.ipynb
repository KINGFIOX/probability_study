```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## 实验1

```python
n = np.zeros(1000)
for i in range(1000):
    n[i] = 0
    pd = np.zeros(100)
    while sum(pd) < 100:
        index = np.arange(100)
        np.random.shuffle(index)
        pd[index[1]] = 1
        n[i] = n[i] + 1

m = np.mean(n)
print(m)
```

```python
arr = np.array([0, 0, 0, 1, 1, 1, 2, 2, 3, 3])
```

```python
epochs = [50, 200, 2000, 5000]
for epoch in epochs:
    sum = 0
    for i in range(epoch):
        np.random.shuffle(arr)  # 洗牌
        sum += arr[0]
    print(sum / epoch)
```

## 计算二项分布

```python

score = [0, 15, 30, 55, 100]
p = 0.6
n = len(score)
```

```python
e = 0
for i in range(n):
    e += math.comb(n - 1, i) * ((1 - p) ** (n - 1 - i)) * (p ** i) * score[i]

print(e)
```
