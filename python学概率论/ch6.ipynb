```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 70, 0.01)  # 0到70的卡方分布曲线，自由度为4
plt.plot(x, stats.chi2.pdf(x, df=4))
plt.plot(x, stats.chi2.pdf(x, df=15))
plt.plot(x, stats.chi2.pdf(x, df=45))
```

```python
x = np.arange(-5, 5, 0.01)
plt.plot(x, stats.t.pdf(x, df=4))
plt.plot(x, stats.t.pdf(x, df=15))
plt.plot(x, stats.t.pdf(x, df=45))
```

```python
x = np.arange(0, 70, 0.01)  # 0到70的卡方分布曲线，自由度为4
plt.plot(x, stats.f.pdf(x, 2, 5))
plt.plot(x, stats.f.pdf(x, 10, 5))
plt.plot(x, stats.f.pdf(x, 100, 5))
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
m, n, mu, sig = 1000, 10, 2, 4
```

```python
xr = np.random.normal(mu, sig, [m, n])  # 正太分布，取样
```

```python
xbar = np.mean(xr, axis=1)
xbar
```

```python
plt.hist(xbar, 20, range=None, density=1)  # 每一格统计一次
# 20 是直方图的柱子数量
# range=None表示自动确定范围
# density=1表示直方图的纵轴是概率密度函数，如果没有设置，则是统计量
```

```python
x2 = np.arange(-5, 5, 0.01)
```

```python
sig1 = sig / np.sqrt(n)
```

```python
plt.plot(x2, stats.norm.pdf(x2, mu, sig1))
```
