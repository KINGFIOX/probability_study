```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n = 1000
x3 = np.linspace(0, 2, 20)  # 0到2之间等差的20份数据
x = np.random.uniform(0, 1, n)  # 均匀分布
y = np.random.uniform(0, 1, n)
z = x + y
```

```python
plt.hist(z, len(x3))
```

```python
x1 = np.linspace(0, 1, 100)
gz1 = x1
x2 = np.linspace(1, 2, 100)
gz2 = 2 - x2
```

```python
plt.plot(x1, gz1 * (n * 0.1), "-")
plt.plot(x1, gz2 * (n * 0.1), "-")
```
